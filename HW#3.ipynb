## **Задача 1** 
1.1 Скачать данные о недвижимости в закрепе к уроку.
Считать данные с помощью pandas
Вывести на экран первые 5 строк




```python
import pandas as pd
import numpy as np

df = pd.read_csv('Downloads/02 Studying/08 Statistic/02 Python/python_for_statistics/kc_house_data.csv')

df.head(5)
```

1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0



```python
df['delta_renovated'] = np.where(df['yr_renovated'] != 0, df['yr_renovated'] - df['yr_built'], np.where(df['yr_renovated'] == 0, 0, None))

df.head()
```

1.3 Создайте признаки года продажи, месяца продажи

```python
df['yr_sell'] = df['date'].str[:4]
df['mnth_sell'] = df['date'].str[4:6]

df.head()
```

1.4 Удалите признаки date, zipcode, lat, long

```python
df = df.drop(columns='date')
df = df.drop(columns='zipcode')
df = df.drop(columns='lat')
df = df.drop(columns='long')

df.head()
```

## **3 задача**

3.1 Найдите среднюю стоимость домов в зависимости от количества спален и отсортируйте от меньшей стоимости к большей

```python
mean_prices = df.groupby('bedrooms')['price'].mean().sort_values()


mean_prices
```

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
house_condition = df.pivot_table(index='grade', aggfunc={'price': ['min','mean','max']})

house_condition
```

3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
water_front_view = pd.pivot_table(df, values='id', index='waterfront', columns='grade', aggfunc='count')

water_front_view.head()
```

3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
groupped = df.groupby(['floors', 'bedrooms'])['id'].count().sort_values(ascending=False)

groupped.head(10)
```

3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
median_price = pd.pivot_table(df, values='price', index='grade', aggfunc='median')

median_price
```
