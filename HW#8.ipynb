```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('AB_res.xlsx')

total_players = data.groupby('version')['userid'].count()
avg_gamerounds = data.groupby('version')['sum_gamerounds'].mean()

retention_1 = data['retention_1'].mean() * 100
retention_7 = data['retention_7'].mean() * 100

print("Результаты эксперимента:")
print("Общее количество игроков в каждой группе:")
print(total_players)
print("Среднее количество сыгранных игр в каждой группе:")
print(avg_gamerounds)
print("Процент удержания игроков на следующий день: {:.2f}%".format(retention_1))
print("Процент удержания игроков через 7 дней: {:.2f}%".format(retention_7))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data[data['version'] == 'gate_30']['sum_gamerounds'], bins=30, alpha=0.5, label='gate_30')
plt.hist(data[data['version'] == 'gate_40']['sum_gamerounds'], bins=30, alpha=0.5, label='gate_40')
plt.xlabel('Количество игр')
plt.ylabel('Частота')
plt.title('Распределение количества сыгранных игр')
plt.legend()
plt.show()

labels = ['Удержание на следующий день', 'Удержание через 7 дней']
values = [retention_1, retention_7]

plt.figure(figsize=(6, 6))
plt.bar(labels, values)
plt.ylabel('Процент')
plt.title('Процент удержания игроков')
plt.show()
```

```python
print("Рекомендация:")
if avg_gamerounds['gate_30'] > avg_gamerounds['gate_40']:
    print("1. Рассмотрите возможность использования порога в 40 уровней, так как среднее количество сыгранных игр выше в этой группе.")
else:
    print("1. Оставайтесь при пороге в 30 уровней, так как среднее количество сыгранных игр не отличается существенно между группами.")

if retention_7 > retention_1:
    print("2. Сосредоточьтесь на удержании игроков через 7 дней после установки, так как процент удержания выше.")
else:
    print("2. Сосредоточьтесь на удержании игроков на следующий день после установки, так как процент удержания выше.")
```
